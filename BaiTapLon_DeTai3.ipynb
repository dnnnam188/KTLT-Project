# BTL03 *[Giáo dục & Học vụ]*

## Hệ thống quản lý điểm và phân tích học lực sinh viên

### Mô tả bài toán

Mỗi sinh viên có danh sách điểm nhiều môn học qua nhiều kỳ. Chương trình tính GPA, phân loại học lực, tìm đoạn kỳ học tiến bộ liên tiếp và phát hiện sinh viên có xu hướng sụt giảm.

### Thiết kế Struct lồng nhau

```cpp
struct Grade   { string subject; double score; int credit; string semester; }
struct Student { string id, name; vector<Grade> grades; double gpa; string rank; }
struct Cohort  { string class_id; vector<Student> students; }
```

→ Cohort → Student → Grade: struct lồng nhau cấp 3

### Input / Output

- **Input:** `grades.csv` (student_id, name, subject, score, credit, semester); `class.txt`
- **Output:** `gpa_report.csv`, `progress.txt` (đoạn tiến bộ), `ranking.csv`

### Yêu cầu kỹ thuật

| Kỹ thuật | Yêu cầu cụ thể |
|---|---|
| Struct lồng nhau | Cohort chứa `vector<Student>`, Student chứa `vector<Grade>`; GPA tính từ `grades[]` |
| File/CSV | Đọc và nhóm Grade vào đúng Student; validate điểm [0,10], tín chỉ > 0 |
| Dãy con – Kadane | Trên mảng GPA từng kỳ của SV: tìm đoạn kỳ có tổng GPA cao nhất |
| Dãy con – tăng dài nhất | Tìm chuỗi kỳ học GPA tăng liên tiếp dài nhất (phát hiện tiến bộ bền vững) |
| Mảng con – Sliding window | Cửa sổ k kỳ: tính GPA trung bình trượt, phát hiện đoạn sụt giảm |
| Prefix sum 1D | Prefix sum tín chỉ tích lũy theo kỳ: truy vấn tổng tín chỉ đoạn [l,r] O(1) |
| Thống kê mô tả | mean/std/min/max điểm theo môn, theo lớp; chuẩn hoá Z-score |
| Xử lý chuỗi | Chuẩn hoá tên SV (trim, title case), phân loại học lực theo chuỗi rank |
| M0 + Test | ≥3 bộ test; case biên: SV 0 môn, điểm toàn 0, 1 kỳ học |
